```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x_range = np.linspace(0.1, 100, 1000)

T_list = [1, 2, 5]
color_list = ['blue', 'green', 'orange', 'red']

for i, T in enumerate(T_list):

    xi = 2*np.pi*T
    beta = 1/T

    def G_ap(x):
        exp_ = np.exp(-x/xi)
        return (1/(beta*np.pi)) * exp_ / (1-exp_**2)
    def G_p(x):
        exp_ = np.exp(-x/xi)
        return (1/(beta*np.pi)) * 1 / (1-exp_**2)

    plt.plot(x_range, G_ap(x_range), color=color_list[i], label=fr"$T = {T} (anti-periodic)$")
    plt.plot(x_range, G_p(x_range), '--', color=color_list[i], label=fr"$T = {T} (periodic)$")

# plt.plot(x_range, 1/(2*np.pi*x_range), "--", color='black', label=r"$T=0$")

plt.plot(x_range, np.ones_like(x_range), ':', color='black')
# plt.xscale('log')
plt.yscale('log')
plt.ylim(1e-7, 1e3)
plt.legend()
plt.savefig("thermal_massless_corr.png")
```

```python
L = 32
m = 0.3

x_range = np.arange(0, L)
G = lambda x : np.sqrt(x*m)*np.exp(-2*x*m)/(2*np.sqrt(np.pi))
G_app = lambda x : np.sqrt(m)*np.exp(-2*x*m)/(2*np.sqrt(x*np.pi))

plt.plot(x_range, G(x_range) + G(L-1-x_range), '.-')
plt.plot(x_range, G_app(x_range) + G_app(L-1-x_range), '.-')
plt.yscale('log')
```
